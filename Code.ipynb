```python
""" Return a Dataframe in columns and rows of the CSV file""" 
# Reading Data
import pandas as pd # we need to improt pandas library
stations = pd.read_csv('ClimateDaily_2023.csv')#reading the csv file. The data is stored in a variable named stations. Calling pd pandas and there is a 
                                               #function in pandascalled read_csv which allow us to read the data of the csv file.

stations
                                                                                        

                                                                                    
```

```python
"""Return a table showing the data with the new geometry column"""
#converting the csv file into shapefile
import geopandas as gpd # we need to import geopandas library 
stations_shp = gpd.GeoDataFrame(stations, crs = 'EPSG:4326', 
                                geometry = gpd.points_from_xy(stations['LONGITUDE'], stations['LATITUDE']))
#stations_shp is the variable where we are going to save the file.
#geometry will be the new column and  we are going to use geopandas library and call geodataframe.
#The CSV file is stored in the stations variable.Then we need to define the epsg projection 
#and we need to specify the longitude and latitude points.
stations_shp
```

```python
#saving the file
stations_shp.to_file(r'stations_shp/stations_shp.shp') # to save the file you will need to create a folder named stations_shp.This shapefile can be used to pluggin Qgis or Arcgis
```

```python
import geopandas as gpd # we need to import geopandas library
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1,figsize = (9,7)) # to create a figure
stations_shp.plot(ax=ax, color = 'g') #specify the axis and the color
plt.show()    
"""Retrurn a simple map showing the place of the climate station as green points"""
```
